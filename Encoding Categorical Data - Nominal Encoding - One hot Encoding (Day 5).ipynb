```python
X=df.iloc[:,0:4]
y=df.selling_price
```

```python
X.shape,y.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
OHE=OneHotEncoder(drop='first',sparse=False,dtype=np.int32)
```

```python
X_train_new=OHE.fit_transform(X_train[['fuel','owner']])
```
